```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r'flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# make 3 separate date , month , year int feature from 24/03/2019

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Date'] = df['Date'].astype(int)

df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Month'] = df['Month'].astype(int)

df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df['Year'] = df['Year'].astype(int)
```

```python
df.head()
```

```python
df.drop('Date_of_Journey',inplace=True,axis=1)
```

```python
df.head()
```

```python
# clean arrival time into arrival hour and mins

df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)

df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1].str[0:2] # or we can use apply method
df['Arrival_Min'] = df['Arrival_Min'].astype(int)
df.head()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'] = df['Total_Stops'].map(
    {'non-stop': 0,
     '2 stops': 1,
     '1 stop': 2,
     '3 stops': 3,
     np.nan: 1,
     '4 stops': 5


     }
)
```

```python
df.head()
```

```python
df['Dep_Time_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Time_Hour'] = df['Dep_Time_Hour'].astype(int)

df['Dep_Time_Min'] = df['Dep_Time'].str.split(':').str[1]

df['Dep_Time_Min'] = df['Dep_Time_Min'].astype(int)

df.drop('Dep_Time', axis=1, inplace=True)


df.head()
```

```python
# cleaning duration 

df['Duration'] = df['Duration'].apply(str.strip)
df['h'] = df['Duration'].str.split(" ").str[0].str.split("h").str[0]

df.loc[df['h'] == '5m', 'h'] = '5' # cleaning one 5m garbage data

df['m'] = df['Duration'].str.split(" ").str[1].str.split("m").str[0]
# df['h'] = df['h'].astype(int)|
df['m'].fillna(0,inplace=True)
df['h'].fillna(0,inplace=True)
df['m'] = df['m'].astype(int)
df['h'] = df['h'].astype(int)


df.head()
```

```python
df['Duration'] = df['h']* 60 + df['m']
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(['h','h'],inplace=True,axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

cat_col = ['Airline','Source','Destination','Additional_Info']
encoded = encoder.fit_transform(df[cat_col])
df2 = pd.DataFrame(encoded,columns= encoder.get_feature_names_out())
df2
```

```python
df_final = pd.concat([df.drop(cat_col, axis=1),df2],axis=1)
df_final.head()
```
