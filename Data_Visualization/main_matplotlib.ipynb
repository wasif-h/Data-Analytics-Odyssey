```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,15)
y = x ** 2
print(x,y,sep="\n\n")
```

```python
# line graph 

plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
plt.subplot(2,2,1)
plt.plot(x**2,y)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(2*x,2*y)
```
