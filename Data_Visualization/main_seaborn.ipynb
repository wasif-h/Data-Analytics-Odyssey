# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head(10)
```

```python
# size is categorical 
df['size'].unique()
```

### Distribution or Histogram plot 

```python
sns.displot(df['total_bill'])
```

```python
sns.displot(df['total_bill'],bins=10, kde=True)
# we can know - the distribution is left skewed
```

```python
# tips and total bill together

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],kde = True)
```

### Join Plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='scatter')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind='hex')
```

### Pairplot

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
# rugplot 

sns.rugplot(df['total_bill'])
```

### Categorical Plot 

```python
sns.countplot(x = df['sex'])
```

```python
sns.countplot(x = df['sex'],hue=df['smoker'])
```

```python
# x = categorical , y = quantative 
sns.barplot(x= df['sex'],y = df['total_bill'])
```

```python
sns.barplot(x= df['sex'],y = df['total_bill'],estimator='sum',hue=df['smoker'])
```

```python
# boxplot 

sns.boxplot(data = df, x = 'tip', y = 'day',palette='rainbow')
```

```python
sns.boxplot(data=df, x = 'day', y = 'total_bill', palette='GnBu')
```

Striplot and swarmplot

```python
sns.swarmplot(x = 'day', y = 'total_bill',data = df, palette='Greys')
sns.boxplot(x = 'day', y = 'total_bill',data = df,palette='vlag')
```

### Matrix Plot

HeatMap

```python
fly = sns.load_dataset("flights")
fly.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_corr = tips[['total_bill','tip','size']]
tips_corr
```

```python
# corelation 
tips_corr.corr()
```

```python
sns.heatmap(tips_corr.corr(),annot=True)
```
