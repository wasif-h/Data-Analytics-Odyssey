### Analysis Questions

```python
# highest population 

df[df['population'] ==  df['population'].max()]['country']
```

```python
# capital of that country
# highest population 

df[df['population'] ==  df['population'].max()]['capital_city']
```

```python
# least population 
df[df['population'] == df['population'].min()]['country']
```

```python
# least population city
df[df['population'] == df['population'].min()]['capital_city']
```

```python
# highest democracy score 

df.sort_values(by='democracy_score',ascending=False)['country'].head()
```

```python
# total region 

df['region']
```

```python
df['region'].value_counts()
```

```python
df['region'].value_counts().count()
```

```python
# number of countries in eastern europe 

df['region'].value_counts()['Eastern Europe']

```

```python
df[df['region'] == 'Eastern Europe']['country']
```

```python
df.sample(10)
```

```python
# 2nd largest population - political leader

df['population'].nlargest(2)
```

```python
df[df['population'] == df['population'].nlargest(2).iloc[1]]
# nlargest[n] and iloc[n - 1]
```

```python
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
# how many unknown political leader / total null values at political col 

df[df['political_leader'].isna()]
```

```python
df[df['political_leader'].isna()]['country'].count()
```
